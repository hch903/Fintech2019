```python
import pandas as pd
import time
import datetime
import xlwt
```

```python
raw_stock_data_2017 = pd.read_excel("stock_data.xlsx", sheet_name="上市2017")
raw_stock_data_2016 = pd.read_excel("stock_data.xlsx", sheet_name="上市2016")
df_2017 = pd.DataFrame(raw_stock_data_2017)
df_2016 = pd.DataFrame(raw_stock_data_2016)
```

```python
df_tsmc = pd.DataFrame(columns=['證券代碼','年月日','開盤價(元)','收盤價(元)'])
j = 0
for index, row in df_2017.iterrows():
    if row['證券代碼'] == '2330 台積電':
        df_tsmc.loc[j] = row
        j += 1
for index, row in df_2016.iterrows():
    if row['證券代碼'] == '2330 台積電':
        df_tsmc.loc[j] = row
        j += 1
```

```python
df_foxconn = pd.DataFrame(columns=['證券代碼','年月日','開盤價(元)','收盤價(元)'])
j = 0
for index, row in df_2017.iterrows():
    if row['證券代碼'] == '2317 鴻海':
        df_foxconn.loc[j] = row
        j += 1
for index, row in df_2016.iterrows():
    if row['證券代碼'] == '2317 鴻海':
        df_foxconn.loc[j] = row
        j += 1
```

```python
df_fpc = pd.DataFrame(columns=['證券代碼','年月日','開盤價(元)','收盤價(元)'])
j = 0
for index, row in df_2017.iterrows():
    if row['證券代碼'] == '1301 台塑':
        df_fpc.loc[j] = row
        j += 1
for index, row in df_2016.iterrows():
    if row['證券代碼'] == '1301 台塑':
        df_fpc.loc[j] = row
        j += 1
```

```python
tsmc_up_date = list()
tsmc_down_date = list()
tsmc_flat_date = list()

# 定義上漲/下跌的幅度 = 1.5%
benchmark = 0.015

for index, row in df_tsmc.iterrows():
    delta = (float(row['收盤價(元)']) - float(row['開盤價(元)'])) / float(row['開盤價(元)'])
    date = row['年月日'].date()
    if delta >= benchmark:
        tsmc_up_date.append(date)
    elif delta <= -1 * benchmark:
        tsmc_down_date.append(date)
    else:
        tsmc_flat_date.append(date)
```
