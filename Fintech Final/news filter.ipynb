```python
import csv
import pandas as pd
import time
```

```python
# 讀檔
raw_news = pd.read_csv("news.csv", encoding="big5")
```

```python
df = pd.DataFrame(raw_news)
df
```

```python
# 各公司關鍵字詞
tsmc = ["台積電", "張忠謀", "劉德音", "魏哲家", "晶圓", "代工", "奈米", "製程", "半導體", "晶片", "積體", "電路"]
foxconn = ["鴻海", "夏普", "郭台銘", "群創", "富士康", "半導體", "通訊元件", "和碩", "友達", "聯華", "蘋概股"]
fpc = ["台塑", "王永慶", "林健男", "王文淵", "王瑞華", "塑膠", "煉油", "石化", "纖維", "紡織", "能源"]
paradox_a = ['買進','賣出']
paradox_b = ['上漲','下跌']
paradox_c = ['跌破','衝上']
paradox_d = ['買超','賣超']
```

```python
df_tsmc = pd.DataFrame(columns=['id','p_type','s_name','s_area_name','post_time','title','author','content','page_url'])
j = 0

for index, row in df.iterrows():
    for i in range(len(tsmc)):
        if (tsmc[i] in str(row['title']) or "台積電" in str(row['content'])) and (row['post_time'] < '2018/1/1'):
            if (paradox_a[0] in str(row['content']) and paradox_a[1] in str(row['content'])):
                continue
            elif (paradox_b[0] in str(row['content']) and paradox_b[1] in str(row['content'])):
                continue
            elif (paradox_c[0] in str(row['content']) and paradox_c[1] in str(row['content'])):
                continue
            elif (paradox_d[0] in str(row['content']) and paradox_d[1] in str(row['content'])):
                continue
            else:
                df_tsmc.loc[j] = row
                j += 1
                break
```

```python
df_foxconn = pd.DataFrame(columns=['id','p_type','s_name','s_area_name','post_time','title','author','content','page_url'])
j = 0
for index, row in df.iterrows():
    for i in range(len(foxconn)):
        if (foxconn[i] in str(row['title']) or "鴻海" in str(row['content'])) and (row['post_time'] < '2018/1/1'):
            if (paradox_a[0] in str(row['content']) and paradox_a[1] in str(row['content'])):
                continue
            elif (paradox_b[0] in str(row['content']) and paradox_b[1] in str(row['content'])):
                continue
            elif (paradox_c[0] in str(row['content']) and paradox_c[1] in str(row['content'])):
                continue
            elif (paradox_d[0] in str(row['content']) and paradox_d[1] in str(row['content'])):
                continue
            else:
                df_foxconn.loc[j] = row
                j += 1
                break
```

```python
with pd.ExcelWriter("company_news(2016~2017).xlsx") as writer:
    df_tsmc.to_excel(writer, sheet_name="TSMC", columns=['id', 'post_time', 'title', 'content'])
    df_foxconn.to_excel(writer, sheet_name="Foxconn", columns=['id', 'post_time', 'title', 'content'])
    df_fpc.to_excel(writer, sheet_name="FPC", columns=['id', 'post_time', 'title', 'content'])
```
