```python
import numpy as np
import numpy.random as nrand
import pandas as pd
import math
```

```python
def preprocess(list,period):
    oldlist = np.array(list)
    newlist = []
    for i in range(0,len(oldlist),period):
        if(i+period<len(oldlist)):
            e = (oldlist[i+period]-oldlist[i])/oldlist[i]
        newlist.append(e)
    newlist = np.array(newlist)
    return newlist
def norm(raw_list):
    e = np.mean(raw_list)
    std_list = (raw_list-e)/e
    return std_list

df = pd.read_excel("output.xlsx")
week_dict = {}
month_dict = {}
for i in range(2,46):
    mylist = np.array(df.iloc(1)[i-1])
    symbol = mylist[0]
    rev_list = mylist[1:][::-1]
    week_dict[symbol] = preprocess(rev_list,5) # 5 day per week
    month_dict[symbol] = preprocess(rev_list,20)# 4 week per month
week_dict
```

```python
df = pd.DataFrame.from_dict(week_dict)
df
```

```python
etf = []
value = []
for column in df:
    try:
        Sw = df[column].convert_objects(convert_numeric=True).skew(skipna=True)
        Kw = df[column].convert_objects(convert_numeric=True).kurtosis(skipna=True)
        Mw = df[column].convert_objects(convert_numeric=True).mean(skipna=True)
        Vw = df[column].convert_objects(convert_numeric=True).var(skipna=True)
        rfw = math.pow((1+0.02822), 1/52)-1
        if 3*Kw-5*math.pow(Sw,2)-9<=0:
            SR = (Mw - rfw) / math.sqrt(Vw)
            ASSR = SR * math.sqrt(1 + Sw * SR / 3)
            value.append(ASSR)
        else:
            Aw = 3*math.sqrt(3*Kw-4*math.pow(Sw,2)-9)/(Vw*(3*Kw-5*math.pow(Sw,2)-9))
            Bw = 3*Sw/(math.sqrt(Vw)*(3*Kw-5*math.pow(Sw,2)-9))
            nw = Mw-3*Sw*math.sqrt(Vw)/(3*Kw-4*math.pow(Sw,2)-9)
            Deltaw = 3*math.sqrt(Vw)*math.sqrt(3*Kw-5*math.pow(Sw,2)-9)/(Vw*(3*Kw-4*math.pow(Sw,2)-9))
            Phiw = math.sqrt(Aw*Aw-Bw*Bw)

            Lambdaw = 1
            Astarw = 1/Lambdaw*((Bw+Aw*(nw-rfw))/math.sqrt(Deltaw*Deltaw+math.pow(nw-rfw,2)))
#             ASKSR = math.sqrt(2*(Lambdaw*Astarw*(nw-rfw)-Deltaw*(Phiw-math.sqrt(Aw*Aw-math.pow(Bw-Lambdaw*Astarw,2)))))
            value.append(Astarw)
    except:
        continue
etf = list(df)

print(len(value))
```

```python
df1 = pd.DataFrame()
df1['etf'] = etf
df1['value'] = value
df1 = df1.sort_values(by=['value'],ascending=False)
print(df1)
df1 = df1.set_index([pd.Series([i for i in range(1,len(df1['value'])+1)])])
df1.index.name="Ranking"
df1.to_excel("week_ASKSR_rank.xlsx")
```

```python
df1 = pd.DataFrame()
df1['etf'] = etf
df1['value'] = value
df1 = df1.sort_values(by=['value'],ascending=False)
print(df1)
df1 = df1.set_index([pd.Series([i for i in range(1,len(df1['value'])+1)])])
df1.index.name="Ranking"
df1.to_excel("month_ASKSR_rank.xlsx")
```
